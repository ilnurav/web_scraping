```python
import requests  # Импортирует библиотеку для выполнения HTTP-запросов.
from bs4 import BeautifulSoup  # Импортирует библиотеку для парсинга HTML и XML документов.
import os  # Импортирует модуль для взаимодействия с файловой системой.
import gdown  # Импортирует библиотеку для загрузки файлов из Google Drive.

k = 1  # Инициализирует счетчик для нумерации товаров.
cwd_path = os.getcwd()  # Получает текущую рабочую директорию.

for page in range(1, 2):  # Итерируется по страницам (с 1 до 2, т.е., только первая страница).
    url = f'https://2krossovka.ru/katalog/?page={page}'  # Формирует URL для текущей страницы.
    response = requests.get(url)  # Выполняет GET-запрос по указанному URL.

    if response.status_code == 200:  # Проверяет, что запрос выполнен успешно (статус 200).
        soup = BeautifulSoup(response.text, 'html.parser')  # Парсит HTML-код страницы.
        products = soup.find_all('div', class_='caption')  # Находит все элементы с классом 'caption'.

        for product in products:  # Итерируется по каждому найденному продукту.
            title = product.h4.a.text.strip()  # Извлекает название товара и убирает лишние пробелы.
            price = product.p.text.strip()  # Извлекает цену товара и убирает лишние пробелы.
            url_product = product.h4.a['href'].strip()  # Извлекает URL товара.

            print(k, title, price, url_product)  # Выводит информацию о товаре.

            response_product = requests.get(url_product)  # Выполняет GET-запрос по URL товара.

            if response_product.status_code == 200:  # Проверяет успешность запроса для товара.
                soup_product = BeautifulSoup(response_product.text, 'html.parser')  # Парсит HTML-код страницы товара.
                table_info = soup_product.find('table', class_='table')  # Находит таблицу с характеристиками товара.

                info_names = table_info.find_all(attrs={'itemprop': 'name'})  # Извлекает названия характеристик.
                info_values = table_info.find_all(attrs={'itemprop': 'value'})  # Извлекает значения характеристик.

                category = ''  # Инициализирует переменную для категории товара.
                names_list = []  # Список для хранения названий характеристик.
                values_list = []  # Список для хранения значений характеристик.

                for i in range(len(info_names)):  # Итерируется по всем характеристикам.
                    names_list.append(info_names[i].text.strip())  # Добавляет название характеристики в список.
                    values_list.append(info_values[i].text.strip())  # Добавляет значение характеристики в список.

                    if info_names[i].text.strip() == 'Категория':  # Проверяет, является ли характеристика категорией.
                        category = info_values[i].text.strip()  # Сохраняет категорию.

                title_url = ''.join(c for c in title if c.isalnum() or c == '_' or c == ' ')  # Формирует безопасное имя для папки из названия товара.
                os.chdir(cwd_path)  # Возвращается в текущую рабочую директорию.

                path_name = f'{cwd_path}\\2krossovka.ru\\{category}\\{title_url}'  # Формирует полный путь для сохранения информации о товаре.

                if not os.path.exists(path_name):  # Проверяет, существует ли папка по этому пути.
                    os.makedirs(path_name)  # Создает папку, если ее нет.

                os.chdir(path_name)  # Переходит в созданную папку.

                with open('info.txt', 'w', encoding='utf-8') as f:  # Открывает файл для записи информации о товаре.
                    f.write(f'Товар: {title}\n')  # Записывает название товара.
                    f.write(f'{price}\n')  # Записывает цену товара.
                    f.write(f'Характеристики:\n')  # Записывает заголовок для характеристик.

                    for i in range(len(names_list)):  # Итерируется по спискам характеристик и значений.
                        f.write(f'{names_list[i]}: {values_list[i]}\n')  # Записывает каждую характеристику и ее значение.

                images_url = soup_product.find_all('a', class_='thumbnail')  # Находит все ссылки на изображения товара.

                for image in images_url:  # Итерируется по всем найденным изображениям.
                    gdown.download(image['href'].strip(), None, quiet=True)  # Загружает изображение.

            k += 1  # Увеличивает счетчик товаров.
            print()  # Печатает пустую строку для разделения товаров в выводе.
```
